```python
# upload the dataset

import pandas as pd
df = pd.read_csv("Financials.csv")
df.head()
```

```python
# chekcng the type of the columns
df.info()
```

```python
# see what years and years we have
profuniq = sorted(df[' Product '].unique())
yearuniq = sorted(df['Year'].unique())
print(profuniq)
print(yearuniq)
```

```python
# remove charcters form numericla column ti turh them to numerical type data
df[' Profit '] = df[' Profit '].replace('[\$,\.\-]', '', regex=True)
df[' Units Sold '] = df[' Units Sold '].replace('[\$,\.\-]', '', regex=True)
df[' Manufacturing Price '] = df[' Manufacturing Price '].replace('[\$,\.\-]', '', regex=True)
df[' Sale Price '] = df[' Sale Price '].replace('[\$,\.\-]', '', regex=True)
df[' Gross Sales '] = df[' Gross Sales '].replace('[\$,\.\-]', '', regex=True)
df[' Discounts '] = df[' Discounts '].replace('[\$,\.\-]', '', regex=True)
df[' COGS '] = df[' COGS '].replace('[\$,\.\-]', '', regex=True)
df['  Sales '] = df['  Sales '].replace('[\$,\.\-]', '', regex=True)
```

```python
# turning some columns to numerical type
df[' Discounts '] = pd.to_numeric(df[' Discounts '], errors="coerce")
df[' Sale Price '] = pd.to_numeric(df[' Sale Price '], errors="coerce")
df[' COGS '] = pd.to_numeric(df[' COGS '], errors="coerce")
df[' Profit '] = pd.to_numeric(df[' Profit '], errors="coerce")
df[' Gross Sales '] = pd.to_numeric(df[' Gross Sales '], errors="coerce")
df[' Manufacturing Price '] = pd.to_numeric(df[' Manufacturing Price '], errors="coerce")
df['  Sales '] = pd.to_numeric(df['  Sales '], errors="coerce")
df[' Units Sold '] = pd.to_numeric(df[' Units Sold '], errors="coerce")
```

```python
# check if it worked
df.info()
```

```python
# turn date column to date type data
df['Date'] = pd.to_datetime(df['Date'])
```

Business and financial

Change in month to month sales

```python
monthlysales = df.groupby([' Product ', 'Year', 'Month Number'])['  Sales '].sum().reset_index()
monthlysales = monthlysales.sort_values([' Product ', 'Year', 'Month Number'])
monthlysales['previous_month_sales'] = monthlysales.groupby(' Product ')['  Sales '].shift(1)
monthlysales['month_growth'] = ((monthlysales['  Sales '] - monthlysales['previous_month_sales']) / monthlysales['previous_month_sales']) * 100
monthlysales
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 10))
for product in monthlysales[' Product '].unique()[:5]:
    productdata = monthlysales[monthlysales[' Product '] == product]
    productdata['period'] = productdata['Year'].astype(str) + '-' + productdata['Month Number'].astype(str).str.zfill(2)
    plt.plot(productdata['period'], productdata['  Sales '], marker='o', linewidth=2, label=product)
plt.title('monthly sales trends', fontsize=14, fontweight='bold')
plt.xlabel('year-month')
plt.ylabel('sales amount')
plt.legend(bbox_to_anchor=(1.1, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

comparing months of each year

```python
yearbyyear = df.groupby([' Product ', 'Month Number', 'Year'])['  Sales '].sum().reset_index()
yearpivot = yearbyyear.pivot_table(index=[' Product ', 'Month Number'],columns='Year', values='  Sales ').reset_index()
year1 = yearpivot.columns[1]
year2 = yearpivot.columns[2]
yearpivot['year_growth'] = ((yearpivot[year2] - yearpivot[year1]) / yearpivot[year1]) * 100
yearpivot.head()
```

```python
import math

products = yearpivot[' Product '].unique()
nproducts = len(products)

cols = 3
rows = math.ceil(nproducts/cols)

fig, axes = plt.subplots(rows, cols, figsize=(18, 4 * rows))
axes = axes.flatten() if nproducts > 1 else [axes]

for i, product in enumerate(products):
    if i < len(axes):
       productdata = yearpivot[yearpivot[' Product '] == product].sort_values('Month Number')
       axes[i].plot(productdata['Month Number'], productdata[year1],
                    marker='o', linewidth=2, label=f'{year1}', alpha = 0.7)
       axes[i].plot(productdata['Month Number'], productdata[year2],
                    marker='s', linewidth=2, label=f'{year2}', alpha = 0.7)
       axes[i].fill_between(productdata['Month Number'],
                            productdata[year1], productdata[year2],
                            alpha=0.2, color='Blue')
       axes[i].set_title(f'{product.strip()}\nTotal Growth: {productdata["year_growth"].mean():.1f}%')
       axes[i].set_xlabel('Month')
       axes[i].set_ylabel('Sales')
       axes[i].legend()
       axes[i].grid(True, alpha=0.4)
       axes[i].set_xticks(range(1,13))

for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

plt.suptitle(f'Sales Comparison: {year1} vs {year2} - All Products',
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

comparing sales in years

```python
yearsales = df.groupby([' Product ', 'Year'])['  Sales '].sum().reset_index()
yeargrowth = yearsales.pivot_table(index=' Product ',
                                  columns = 'Year', values='  Sales ').reset_index()
year1 = yeargrowth.columns[1]
year2 = yeargrowth.columns[2]
yeargrowth['yeargrowthpct'] = ((yeargrowth[year2] - yeargrowth[year1]) / yeargrowth[year1]) * 100
yeargrowth
```

```python
plt.figure(figsize=(16,12))
plt.bar(yeargrowth[' Product '], yeargrowth['yeargrowthpct'], color='purple')
plt.title('yearly growth percentage', fontsize=14, fontweight='bold', color ='Black')
plt.xlabel('product')
plt.ylabel('growth percentage')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=50)
plt.tight_layout()
plt.show()
```

# Sales performance

Finding how many new orders

```python
neworders = df.groupby(['Date', 'Country', 'Segment']).size().reset_index(name='new_orders')

plt.figure(figsize=(15,9))
for Segment in neworders['Segment'].unique():
    Segmentdata = neworders[neworders['Segment'] == Segment]
    plt.plot(Segmentdata['Date'], Segmentdata['new_orders'], label=Segment, marker='o')
plt.title('new orders by product segment')
plt.legend()
plt.xticks(rotation=50)
plt.show()
```

# checking how many new customers

```python
newcustomers = df.groupby([df['Date'].dt.to_period('Y'), 'Country'])['  Sales '].count().reset_index(name='new_customers')

plt.figure(figsize=(14,8))
newcustomers.pivot(index='Date', columns='Country', values='new_customers').plot(kind='bar')
plt.title('new orders by year and country')
plt.legend()
plt.show()
```

```python
# average profit by segment
salesperformance = df.groupby(['Country', 'Segment']).agg({'  Sales ': 'sum',
                                                          ' Profit ':'sum',
                                                          ' Units Sold ': 'sum',
                                                          ' Discounts ': 'sum'}).reset_index()
salesperformance['profitmargin'] = (salesperformance[' Profit '] / salesperformance['  Sales ']) * 100

plt.figure(figsize=(14,8))
salesperformance.groupby('Segment')['profitmargin'].mean().plot(kind='bar', color='blue')
plt.title('avg profit by segment')
plt.ylabel('profit margin %')
plt.show()
```

```python
# relation between discount and profit
salesperformance['avgtdiscountrate'] = (salesperformance[' Discounts '] / salesperformance['  Sales ']) * 100

plt.figure(figsize=(14,8))
plt.scatter(salesperformance['avgtdiscountrate'], salesperformance['profitmargin'], alpha=0.4, color='red')
plt.xlabel('avg disocunt rate')
plt.ylabel('profit margin %')
plt.title('Discount/Profit rates')
plt.grid(True)
```
